# Brute Force Algorithm


## Finding the Maximum Value in an Array

In this notebook, we'll explore a brute-force algorithm to find the maximum value within an array of integers.

**Brute-Force Approach**

The brute-force method involves comparing each element of the array against a variable that keeps track of the current maximum value. Here's the algorithm in pseudocode:

```
Algorithm 2: Finding Maximum of an Array of Integers

1: procedure Max(A[1::n])
2:    max <- a1
3:    for i = 2 to n do
4:        if ai > max then
5:            max <- ai
6:        end if
7:    end for
8: end procedure
```
**Explanation:**

1. We initialize a variable `max` with the first element of the array `A`.
2. We iterate through the array from the second element to the last.
3. Inside the loop, we compare each element (`ai`) with the current `max`.
4. If an element is greater than `max`, we update `max` with that element's value.
5. After the iteration completes, the `max` variable holds the maximum value in the array.

````python
import matplotlib.pyplot as plt
import numpy as np
import timeit

def find_max(A):
    max_value = A[0]
    for i in range(1, len(A)):
        if A[i] > max_value:
            max_value = A[i]
    return max_value

# Input sizes to test
input_sizes = range(100, 10000, 100)
execution_times = []

# Measure execution time for each input size
for size in input_sizes:
    data = np.random.randint(1, 1000, size=size)  # Generate random data
    time_taken = timeit.timeit(lambda: find_max(data), number=5) / 5  # Average over 5 runs
    execution_times.append(time_taken)

plt.plot(input_sizes, execution_times)
plt.xlabel("Input Size")
plt.ylabel("Execution Time (seconds)")
plt.title("Time Complexity of Brute-Force Max Finding Algorithm")
plt.grid(True)
plt.show()
````

## Sequential Search

**Problem:**  Given an array of integers `A` and an element `x`, determine if `x` exists within the array. If found, return its index; otherwise, return -1.

**Brute-Force Approach (Sequential Search):**

Iterate through the array, comparing each element with `x`.

```
Algorithm 4: Finding an Element in an Array of Integers

1: procedure SeqSearch(A[1::n]; x)
2:    i <- 1
3:    while i < n and a[i] != x do
4:        i <- i + 1
5:    end while
6:    if a[i] = x then
7:        idx <- i
8:    else
9:        idx <- -1
10:   end if
11:   return idx
12: end procedure

```





````python
import matplotlib.pyplot as plt
import numpy as np
import timeit

def SeqSearch(A, x):
    i = 1
    while i < len(A) and A[i-1] != x:
        i += 1

    if A[i-1] == x:
        return i
    else:
        return -1

# Input sizes to test
input_sizes = range(100, 10000, 100)
execution_times = []

# Measure execution time for each input size
for size in input_sizes:
    data = np.random.randint(1, 1000, size=size)
    target = np.random.choice(data)  # Target is now part of the data

    time_taken = timeit.timeit(lambda: SeqSearch(data, target), number=5) / 5
    execution_times.append(time_taken)

# Adjusted Theoretical O(n) Trend Line
theoretical_times = [0.00005 * size for size in input_sizes]

plt.plot(input_sizes, execution_times, label="Empirical Time")
plt.plot(input_sizes, theoretical_times, linestyle='--', color='red', label="O(n) Trend")
plt.xlabel("Input Size")
plt.ylabel("Execution Time (seconds)")
plt.title("Time Complexity of Sequential Search")
plt.grid(True)
plt.legend()
plt.show()
````

## Computing Powers

**Problem:** Calculate the value of `a^n` where `a` is a positive number, and `n` is a non-negative integer.

**Brute-Force Approach:**  Multiply `a` by itself `n` times.

```
Algorithm 6: Computing Powers

1: procedure Power(a; n)
2:    result <- 1
3:    for i = 1 to n do
4:        result <- result * a
5:    end for
6:    return result
7: end procedure
```

**5. (Optional) Time Complexity Analysis**

You can add additional code cells to analyze the time complexity (which is O(n) due to the single loop) and plot the results for varying input sizes. Refer to the previous examples for details on how to do this.

````python
import matplotlib.pyplot as plt
import numpy as np
import timeit

def computingPowers(A, x):
    result = 1
    for i in range(x):
        result *= A
    return result

# Input sizes to test
input_sizes = range(1, 20)  # Exponents can get large quickly
execution_times = []

# Measure execution time for each input size
for size in input_sizes:
    base = np.random.randint(2, 5)  # Keep bases moderate
    time_taken = timeit.timeit(lambda: computingPowers(base, size), number=5) / 5
    execution_times.append(time_taken)

# Theoretical O(n) line
theoretical_times = [0.005 * size for size in input_sizes]

plt.plot(input_sizes, execution_times, label="Empirical Time")
plt.plot(input_sizes, theoretical_times, linestyle='--', color='red', label="O(n) Trend")
plt.xlabel("Exponent (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Time Complexity of Computing Powers")
plt.grid(True)
plt.legend()
plt.show()
````

## Matrix Multiplication

**Problem:**  Given two square matrices A and B of size n x n, compute their product C.

**Brute-Force Approach:** Calculate each element `cij` of the result matrix C by iteratively multiplying the corresponding elements of row `i` of matrix A and column `j` of matrix B and summing the products.

```
Algorithm 9: Square Matrix Multiplication
1: procedure MatrixMult(A; B)
2:    for i <- 1 to n do
3:        for j <- 1 to n do
4:            C[i ; j ] <- 0
5:            for k <- 1 to n do
6:                C[i ; j ] <- C[i ; j ] + A[i ; k] * B[k; j ]
7:            end for
8:        end for
9:    end for
10:   return C
11: end procedure
```

**(Optional) Time Complexity Analysis**

Add code cells to analyze the algorithm's time complexity (O(n^3) due to the nested loops) and plot the execution time for different input sizes.

````python
import matplotlib.pyplot as plt
import numpy as np
import timeit

def MatrixMult(A, B):
    """
    Implements matrix multiplication using a brute-force approach.

    Args:
        A: A square matrix of size n x n.
        B: A square matrix of size n x n.

    Returns:
        The resulting product matrix C of size n x n.
    """
    n = len(A)
    C = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i, j] += A[i, k] * B[k, j]

    return C

# Input sizes to test
input_sizes = range(10, 100, 5)  # Adjust the range for matrix multiplication
execution_times = []

# Measure execution time for each input size
for size in input_sizes:
    # Generate two random matrices
    A = np.random.randint(1, 10, size=(size, size))
    B = np.random.randint(1, 10, size=(size, size))

    time_taken = timeit.timeit(lambda: MatrixMult(A, B), number=5) / 5
    execution_times.append(time_taken)

# Theoretical O(n^3) line (adjust the scaling factor)
theoretical_times = [0.000005 * size**3 for size in input_sizes]

plt.plot(input_sizes, execution_times, label="Empirical Time")
plt.plot(input_sizes, theoretical_times, linestyle='--', color='red', label="O(n^3) Trend")
plt.xlabel("Input Size (n)")  # Note: n is the matrix dimension
plt.ylabel("Execution Time (seconds)")
plt.title("Time Complexity of Matrix Multiplication")
plt.grid(True)
plt.legend()
plt.show()
````

## Prime Number Checking

**Problem:**  Given a positive integer 'n', determine if it is a prime number.

**Definition:** A prime number (greater than 1) is only divisible by 1 and itself.

**Approaches:**

* **Brute-Force:**  Test divisibility of 'n' by numbers from 2 up to n-1.
* **Sieve of Eratosthenes (Optimized):**  Use a more efficient precomputation-based technique to identify prime numbers. We'll focus on the implementation of checking primality using this method.

```
Algorithm 10: Prime Number Test (Brute-Force)

1. If 'n' is less than 2, it's not prime.
2.  Set  `isprime` to True, and an iterator 'k' to 2.
3.  While `isprime` is True and 'k' is less than `n-1`:
    *   If 'n' is divisible by 'k', set `isprime` to False.
    *   Otherwise, increment 'k'.
4.  Return `isprime`
```
```
Algorithm 11: Prime Number Test (Sieve of Eratosthenes Optimization)

1. If 'n' is less than 2, it's not prime.
2. Set `isprime` to True and an iterator 'k' to 2.
3. While `isprime` is True and  'k' squared is less than or equal to 'n':
    * If 'n' is divisible by 'k', set `isprime` to False.
    * Otherwise, increment 'k'.
4. Return `isprime`
```


````python
import matplotlib.pyplot as plt
import numpy as np
import timeit

def IsPrime(n):
    # Brute-Force
    if n <= 1:
        # Numbers less than or equal to 1 are not prime
        return False
    # Check for divisibility from 2 up to n-1
    for i in range(2, n):
        if n % i == 0:
            return False
    # If 'n' is divisible by any number, it's not prime
    return True

def IsPrime2(n):
    # Sieve of Eratosthenes Optimization
    if n <= 1:
        return False
    if n <= 3:
        return True  # 2 and 3 are prime
    if n % 2 == 0 or n % 3 == 0:
        return False  # Handle multiples of 2 and 3 separately
    # Check divisibility by numbers of the form 6k  1
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Input sizes to test (a sample range of numbers)
input_sizes = [11, 13, 17, 23, 31, 43, 59, 71, 89, 101]

execution_times_brute = []
execution_times_sieve = []

for num in input_sizes:
    time_taken = timeit.timeit(lambda: IsPrime(num), number=10) / 10
    execution_times_brute.append(time_taken)

    time_taken = timeit.timeit(lambda: IsPrime2(num), number=10) / 10
    execution_times_sieve.append(time_taken)

# Plotting
plt.plot(input_sizes, execution_times_brute, label="Brute-Force")
plt.plot(input_sizes, execution_times_sieve, label="Sieve Optimization")
plt.xlabel("Input Number ('n')")
plt.ylabel("Execution Time (seconds)")
plt.title("Time Performance of Prime Checking Algorithms")
plt.grid(True)
plt.legend()
plt.show()
````

## Polynomial Evaluation

**Problem:**  Evaluate the polynomial p(x) = a_nx^n + a_{n-1}x^{n-1} + ... + a_1x + a_0 at a given value x = t.

**Approaches:**

* **Brute-Force:** Explicitly calculate each power of 'x' (x^k) and multiply by the corresponding coefficient a_k.
* **Optimized (Horner's Method):**  Uses a more efficient factoring technique to reduce repeated calculations.
```
Algorithm 12 Polynomial interpolation
1: procedure Polynom(n; A[0::n]; t)
2:    p <- 0
3:       for i <- n downto 0 do
4:          power <- 1
5:          for j <- 1 to i do
6:             power <- power * t
7:       end for
8:       p <- p + a[i] * power
9:    end for
10:   return p
11: end procedure


````python
import matplotlib.pyplot as plt
import numpy as np
import timeit

def polynomial_brute_force(A, t):
    n = len(A) - 1  # Degree of the polynomial
    p = 0
    for i in range(n + 1):
        power = 1
        for j in range(i):
            power *= t
        p += A[i] * power
    return p

def polynomial_optimized(A, t):
    n = len(A) - 1
    p = 0
    for i in range(n, -1, -1):
        p = A[i] + t * p
    return p

# Input sizes (degree of the polynomial)
input_sizes = range(5, 50, 5)
execution_times_brute = []
execution_times_optimized = []

for n in input_sizes:
    # Generate random coefficients for a polynomial of degree 'n'
    coefficients = np.random.randint(-10, 10, size=n + 1)
    x_value = 2  # Fixed value for evaluation

    time_taken = timeit.timeit(lambda: polynomial_brute_force(coefficients, x_value), number=5) / 5
    execution_times_brute.append(time_taken)

    time_taken = timeit.timeit(lambda: polynomial_optimized(coefficients, x_value), number=5) / 5
    execution_times_optimized.append(time_taken)

# Plotting (Adjust trend lines if needed)
plt.plot(input_sizes, execution_times_brute, label="Brute-Force")
plt.plot(input_sizes, execution_times_optimized, label="Optimized (Horner's)")

# Add trend lines for Big O notation
plt.plot(input_sizes, [0.00001 * n**2 for n in input_sizes], linestyle='--', color='r', label="O(n^2) Trend")
plt.plot(input_sizes, [0.000001 * n for n in input_sizes], linestyle='--', color='g', label="O(n) Trend")

plt.xlabel("Polynomial Degree")
plt.ylabel("Execution Time (seconds)")
plt.title("Time Complexity of Polynomial Evaluation")
plt.grid(True)
plt.legend()
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
import timeit

def polynomial_brute_force(A, t):
    n = len(A) - 1  # Degree of the polynomial
    p = 0
    for i in range(n + 1):
        power = 1
        for j in range(i):
            power *= t
        p += A[i] * power
    return p

def polynomial_optimized(A, t):
    n = len(A) - 1
    p = 0
    for i in range(n, -1, -1):
        p = A[i] + t * p
    return p

# Input sizes (degree of the polynomial)
input_sizes = range(5, 100, 5)

# Best case
execution_times_brute_best = []
execution_times_optimized_best = []
for n in input_sizes:
    coefficients = np.zeros(n + 1)  # All coefficients are zero except for the constant term
    coefficients[0] = 1
    x_value = 1  # Evaluate at x = 1

    time_taken = timeit.timeit(lambda: polynomial_brute_force(coefficients, x_value), number=5) / 5
    execution_times_brute_best.append(time_taken)

    time_taken = timeit.timeit(lambda: polynomial_optimized(coefficients, x_value), number=5) / 5
    execution_times_optimized_best.append(time_taken)

# Average case
execution_times_brute_avg = []
execution_times_optimized_avg = []
for n in input_sizes:
    coefficients = np.random.randint(-100, 100, size=n + 1)  # Random coefficients
    x_value = 2  # Evaluate at x = 2

    time_taken = timeit.timeit(lambda: polynomial_brute_force(coefficients, x_value), number=5) / 5
    execution_times_brute_avg.append(time_taken)

    time_taken = timeit.timeit(lambda: polynomial_optimized(coefficients, x_value), number=5) / 5
    execution_times_optimized_avg.append(time_taken)

# Worst case
execution_times_brute_worst = []
execution_times_optimized_worst = []
for n in input_sizes:
    coefficients = np.ones(n + 1)  # All coefficients are 1
    x_value = 2  # Evaluate at x = 2

    time_taken = timeit.timeit(lambda: polynomial_brute_force(coefficients, x_value), number=5) / 5
    execution_times_brute_worst.append(time_taken)

    time_taken = timeit.timeit(lambda: polynomial_optimized(coefficients, x_value), number=5) / 5
    execution_times_optimized_worst.append(time_taken)

# Plotting
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 12))

# Best case
ax1.plot(input_sizes, execution_times_brute_best, label="Brute-Force")
ax1.plot(input_sizes, execution_times_optimized_best, label="Optimized (Horner's)")
ax1.set_title("Best Case")
ax1.set_xlabel("Polynomial Degree")
ax1.set_ylabel("Execution Time (seconds)")
ax1.grid(True)
ax1.legend()

# Average case
ax2.plot(input_sizes, execution_times_brute_avg, label="Brute-Force")
ax2.plot(input_sizes, execution_times_optimized_avg, label="Optimized (Horner's)")
ax2.set_title("Average Case")
ax2.set_xlabel("Polynomial Degree")
ax2.set_ylabel("Execution Time (seconds)")
ax2.grid(True)
ax2.legend()

# Worst case
ax3.plot(input_sizes, execution_times_brute_worst, label="Brute-Force")
ax3.plot(input_sizes, execution_times_optimized_worst, label="Optimized (Horner's)")
ax3.set_title("Worst Case")
ax3.set_xlabel("Polynomial Degree")
ax3.set_ylabel("Execution Time (seconds)")
ax3.grid(True)
ax3.legend()

plt.tight_layout()
plt.show()
````
